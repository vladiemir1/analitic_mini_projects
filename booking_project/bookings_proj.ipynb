```python
import pandas as pd
hotel = pd.read_csv(r"C:\Users\Владимир\Desktop\mini_project1\bookings.csv", sep = ';')
```

```python
def lower_underscore(name):
    name = name.lower().replace(' ', '_')  
    return name

hotel = hotel.rename(columns=lower_underscore)
```

```python
hotel['total_kids'] = hotel['children'] + hotel['babies']
```

```python
hotel.groupby('hotel')\
    .agg({'total_kids':'mean'})\
    .round(2)
```

```python
hotel['has_kids'] = hotel['total_kids'] >= 1 
```

```python
hotel.query('has_kids == True and is_canceled == 1').shape[0] /  hotel.query('has_kids == True').shape[0]
```

```python
hotel.query('has_kids == False and is_canceled == 1').shape[0] / hotel.query('has_kids == False').shape[0]
```
